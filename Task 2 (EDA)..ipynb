```python
import matplotlib.pyplot as plt

class_counts = combined_df['topic'].value_counts()

# Create a pie chart
plt.figure(figsize=(7, 7))  # Set the size of the figure (width, height) in inches
plt.pie(class_counts.values, labels=class_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Class Distribution')
plt.show()
```

```python
# Print class count
print("Class Counts:")
for class_label, count in class_counts.items():
    print(f"{class_label}: {count}")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
# Assuming you want to consider trigrams (n=3)
n_gram_range = (3, 3)
top_n = 10

vectorizer = CountVectorizer(ngram_range=n_gram_range, analyzer='word')
ngram_matrix = vectorizer.fit_transform(combined_df['story'])
feature_names = vectorizer.get_feature_names_out()
ngram_freqs = zip(feature_names, ngram_matrix.sum(axis=0).tolist()[0])
sorted_ngrams = sorted(ngram_freqs, key=lambda x: x[1], reverse=True)[:top_n]

print("Top", top_n, "frequent trigrams:")
for ngram, freq in sorted_ngrams:
    print(ngram, ":", freq)
```

```python
top_ngrams_per_class = {}

for class_label, group in combined_df.groupby('topic'):
    ngram_matrix = vectorizer.fit_transform(group['story'])
    feature_names = vectorizer.get_feature_names_out()
    ngram_freqs = zip(feature_names, ngram_matrix.sum(axis=0).tolist()[0])
    sorted_ngrams = sorted(ngram_freqs, key=lambda x: x[1], reverse=True)[:top_n]
    top_ngrams_per_class[class_label] = sorted_ngrams

# Print the top n-grams for each class
for class_label, ngrams in top_ngrams_per_class.items():
    print(f"Top {top_n} frequent trigrams for class '{class_label}':")
    for ngram, freq in ngrams:
        print(ngram, ":", freq)
    print('\n')
```

```python
combined_df['word_count'] = combined_df['story'].apply(lambda x: len(x.split()))
combined_df['letter_count'] = combined_df['story'].apply(lambda x: len(x.replace(" ", "")))

# Plot histogram for word count
plt.hist(combined_df['word_count'], bins=30)
plt.xlabel('Word Count')
plt.ylabel('Frequency')
plt.title('Distribution of Example Lengths in Words')
plt.show()

# Plot histogram for letter count
plt.hist(combined_df['letter_count'], bins=30)
plt.xlabel('Letter Count')
plt.ylabel('Frequency')
plt.title('Distribution of Example Lengths in Letters')
plt.show()
```
